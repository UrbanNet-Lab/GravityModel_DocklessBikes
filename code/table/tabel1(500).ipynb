```python
from math import sin, asin, cos, radians, fabs, sqrt
EARTH_RADIUS=6371 
def hav(theta):
    s = sin(theta / 2)
    return s * s
def get_distance_hav(lat0, lng0, lat1, lng1):    #lat纬度
    lat0 = radians(lat0)
    lat1 = radians(lat1)
    lng0 = radians(lng0)
    lng1 = radians(lng1)
    dlng = fabs(lng0 - lng1)
    dlat = fabs(lat0 - lat1)
    h = hav(dlat) + cos(lat0) * cos(lat1) * hav(dlng)
    distance = 2 * EARTH_RADIUS * asin(sqrt(h))
    return distance
```

```python
pa=(np.load("500mpa(bj).npy")).tolist()
pb=(np.load("500mpb(bj).npy")).tolist()
ww=(np.load("500mww(bj).npy")).tolist()
T_total2=(np.load("500mT(bj).npy")).tolist()
```

```python
dataframe=pd.DataFrame(columns=["Pa","Pb","d","T"])
dataframe["Pa"]=pd.Series(pa)
dataframe["Pb"]=pd.Series(pb)
dataframe["d"]=pd.Series(ww)
dataframe["T"]=pd.Series(T_total2)
```

```python
dataframe1=dataframe[dataframe["Pa"]!=0]   
dr1=dataframe1[dataframe1["Pb"]!=0]
```

```python
a=[]
b=[]
c=[]
r=[]
for _ in range(10000):
    np.random.seed()
    dt = dr1.sample(20000, replace=True)
    pa=(dt.Pa).tolist()
    pb=(dt.Pb).tolist()
    ww=(dt.d).tolist()
    T_total2=(dt["T"]).tolist()
    
    
    pa_=[]
    for i in range(0,len(pa)):
        z=pa[i]
        if (z!=0):
            z1=math.log10(z)
            pa_.append(z1)
        else:
            z1=0
            pa_.append(z1)
    pb_=[]
    for i in range(0,len(pb)):
        z=pb[i]
        if (z!=0):
            z1=math.log10(z)
            pb_.append(z1)
        else:
            z1=0
            pb_.append(z1)         
    ww_=[]
    for i in range(0,len(ww)):
        z=ww[i]
        if (z!=0):
            z1=math.log10(z)
            ww_.append(z1)
        else:
            z1=0
            ww_.append(z1)         
    logT=[]
    for i in range(0,len(T_total2)):
        z=T_total2[i]
        if (z!=0):
            z1=math.log10(z)
            logT.append(z1)
        else:
            z1=0
            logT.append(z1)        
    data=[]
    for i in range(0,len(pa_)):
        c=[]
        z1=pa_[i]
        z2=pb_[i]
        z3=ww_[i]
        #z4=logT[i]
        c.append(z1)
        c.append(z2)
        c.append(z3)
        #c.append(z4)
        data.append(c)        
            
    x=data
    y=logT        
    regr = linear_model.LinearRegression()
    regr.fit(x, y)
    con_a=regr.coef_
    con_b=regr.intercept_
    constant=10**con_b
    a.append(con_a[0])
    b.append(con_a[1])
    r.append(con_a[2])
    c.append(constant)
    
```
