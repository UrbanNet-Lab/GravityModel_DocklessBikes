# 求取每个参数后，求取每个参数的置信区间

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
import copy
import time
import sys
import math
import datetime
from math import radians, cos, sin, asin, sqrt
from matplotlib.colors import LogNorm
from io import StringIO
from sklearn import linear_model
#import tensorflow as tf
import seaborn as sns
from scipy import optimize
import math
#from scipy.optimize import curve_fit
plt.rcParams["font.sans-serif"]=["SimHei"] #用来正常显示中文标签
plt.rcParams["axes.unicode_minus"]=False #用来正常显示负号
```

```python
pa=(np.load("1kmpa(bj).npy")).tolist()
pb=(np.load("1kmpb(bj).npy")).tolist()
ww=(np.load("1kmww(bj).npy")).tolist()
T_total2=(np.load("1kmT(bj).npy")).tolist()
```

```python
dataframe=pd.DataFrame(columns=["Pa","Pb","d","T"])
dataframe["Pa"]=pd.Series(pa)
dataframe["Pb"]=pd.Series(pb)
dataframe["d"]=pd.Series(ww)
dataframe["T"]=pd.Series(T_total2)
```

```python
dataframe1=dataframe[dataframe["Pa"]!=0]   
dr1=dataframe1[dataframe1["Pb"]!=0]
```

```python
a=[]
b=[]
c=[]
r=[]
for _ in range(10000):
    np.random.seed()
    dt = dr1.sample(20000, replace=True)
    pa=(dt.Pa).tolist()
    pb=(dt.Pb).tolist()
    ww=(dt.d).tolist()
    T_total2=(dt["T"]).tolist()
    
    
    pa_=[]
    for i in range(0,len(pa)):
        z=pa[i]
        if (z!=0):
            z1=math.log10(z)
            pa_.append(z1)
        else:
            z1=0
            pa_.append(z1)
    pb_=[]
    for i in range(0,len(pb)):
        z=pb[i]
        if (z!=0):
            z1=math.log10(z)
            pb_.append(z1)
        else:
            z1=0
            pb_.append(z1)         
    ww_=[]
    for i in range(0,len(ww)):
        z=ww[i]
        if (z!=0):
            z1=math.log10(z)
            ww_.append(z1)
        else:
            z1=0
            ww_.append(z1)         
    logT=[]
    for i in range(0,len(T_total2)):
        z=T_total2[i]
        if (z!=0):
            z1=math.log10(z)
            logT.append(z1)
        else:
            z1=0
            logT.append(z1)        
    data=[]
    for i in range(0,len(pa_)):
        c=[]
        z1=pa_[i]
        z2=pb_[i]
        z3=ww_[i]
        #z4=logT[i]
        c.append(z1)
        c.append(z2)
        c.append(z3)
        #c.append(z4)
        data.append(c)        
            
    x=data
    y=logT        
    regr = linear_model.LinearRegression()
    regr.fit(x, y)
    con_a=regr.coef_
    con_b=regr.intercept_
    constant=10**con_b
    a.append(con_a[0])
    b.append(con_a[1])
    r.append(con_a[2])
    c.append(constant)
    
```

```python
print(np.percentile(a, 2.5), np.percentile(a, 97.5))
print(np.percentile(b, 2.5), np.percentile(b, 97.5))
print(np.percentile(r, 2.5), np.percentile(r, 97.5))
print(np.percentile(c, 2.5), np.percentile(c, 97.5))
```
