```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
import copy
import time
import sys
import math
import datetime
from math import radians, cos, sin, asin, sqrt
from matplotlib.colors import LogNorm
from io import StringIO
from sklearn import linear_model
#import tensorflow as tf
from scipy import optimize
import math
#from scipy.optimize import curve_fit
plt.rcParams["font.sans-serif"]=["SimHei"] #用来正常显示中文标签
plt.rcParams["axes.unicode_minus"]=False #用来正常显示负号
```

```python
from math import sin, asin, cos, radians, fabs, sqrt
EARTH_RADIUS=6371 
def hav(theta):
    s = sin(theta / 2)
    return s * s
def get_distance_hav(lat0, lng0, lat1, lng1):    #lat纬度
    lat0 = radians(lat0)
    lat1 = radians(lat1)
    lng0 = radians(lng0)
    lng1 = radians(lng1)
    dlng = fabs(lng0 - lng1)
    dlat = fabs(lat0 - lat1)
    h = hav(dlat) + cos(lat0) * cos(lat1) * hav(dlng)
    distance = 2 * EARTH_RADIUS * asin(sqrt(h))
    return distance
```

通过ArcGis对城市空间进行划分，将城市空间划分成上下左右等距的方形区域，以及区域中心经纬度坐标。然后对每个区域的人口数据进行求和得到区域人口，将获得的每个区域人口与区域id相对应，对每条骑行记录判定其起始/终点所在区域位置，然后统计来往区域对之间的单车骑行流量，并求取两区域中心的距离，得到任意两区域之间的单车流量、两区域人口数据、两区域中心距离。

```python
pa=(np.load("500mpa(bj).npy")).tolist()      #区域人口
pb=(np.load("500mpb(bj).npy")).tolist()      #区域人口
ww=(np.load("500mww(bj).npy")).tolist()      #两区域中心距离
T_total2=(np.load("500mT(bj).npy")).tolist() #两区域单车流量
```

```python
dataframe=pd.DataFrame(columns=["Pa","Pb","d","T"])
dataframe["Pa"]=pd.Series(pa)
dataframe["Pb"]=pd.Series(pb)
dataframe["d"]=pd.Series(ww)
dataframe["T"]=pd.Series(T_total2)
```

```python
dataframe1=dataframe[dataframe["Pa"]!=0]   
dr1=dataframe1[dataframe1["Pb"]!=0]
```

```python
pa=(dr1.Pa).tolist()
pb=(dr1.Pb).tolist()
ww=(dr1.d).tolist()
T_total2=(dr1["T"]).tolist()
```

```python
pa_=[]
for i in range(0,len(pa)):
    z=pa[i]
    if (z!=0):
        z1=math.log10(z)
        pa_.append(z1)
    else:
        z1=0
        pa_.append(z1)
```

```python
pb_=[]
for i in range(0,len(pb)):
    z=pb[i]
    if (z!=0):
        z1=math.log10(z)
        pb_.append(z1)
    else:
        z1=0
        pb_.append(z1)
```

```python
ww_=[]
for i in range(0,len(ww)):
    z=ww[i]
    if (z!=0):
        z1=math.log10(z)
        ww_.append(z1)
    else:
        z1=0
        ww_.append(z1)
```

```python
logT=[]
for i in range(0,len(T_total2)):
    z=T_total2[i]
    if (z!=0):
        z1=math.log10(z)
        logT.append(z1)
    else:
        z1=0
        logT.append(z1)
```

```python
data=[]
for i in range(0,len(pa_)):
    c=[]
    z1=pa_[i]
    z2=pb_[i]
    z3=ww_[i]
    #z4=logT[i]
    c.append(z1)
    c.append(z2)
    c.append(z3)
    #c.append(z4)
    data.append(c)
```

```python
x=data
y=logT
```

```python
regr = linear_model.LinearRegression()
regr.fit(x, y)
 
print ("x的系数值a：")
con_a=regr.coef_
print (regr.coef_)

print ("截距b:")
con_b=regr.intercept_
print (regr.intercept_)

yPred = regr.predict(x)
```

```python
constant=10**con_b
constant
```

```python
alpha_pa=[]  #pa的阿尔法
for i in range(0,len(pa)):
    z=pa[i]**con_a[0]
    alpha_pa.append(z)
```

```python
alpha_pb=[]
for i in range(0,len(pb)):
    z=pb[i]**con_a[1]
    alpha_pb.append(z)
```

```python
alpha_ww=[]
for i in range(0,len(ww)):
    z=ww[i]**-con_a[2]
    alpha_ww.append(z)
```

```python
model=[]
for i  in range(0,len(alpha_pa)):
    z=constant*alpha_pa[i]*alpha_pb[i]/alpha_ww[i]
    model.append(z)
```

```python
model1=[]
for i in range(0,len(model)):
    z=model[i]
    if z!=0:
        
        z1=math.log10(z)
        model1.append(z1)
    else:
        model1.append(0)
```
